```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
from tqdm import tqdm
#import config
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import pdb
import os 
import sys
import glob
import re
import signal_utils

sbp_avgs = pd.read_csv(os.path.join(signal_utils.output_path, "sbp_avgs.csv"), index_col=0)
sbp_avgs.index = pd.to_datetime(sbp_avgs.index)

sbp_avgs['days'] = (sbp_avgs.index - sbp_avgs.index[0]).to_series().dt.days.to_numpy()

sbp_stds = pd.read_csv(os.path.join(signal_utils.output_path, "sbp_stds.csv"), index_col=0)
sbp_stds.index = pd.to_datetime(sbp_stds.index)

sbp_stds['days'] = (sbp_stds.index - sbp_stds.index[0]).to_series().dt.days.to_numpy()
```

```python
for i in range(96): 
    str_c = f'{i}'
    plt.plot(sbp_avgs['days'], sbp_avgs[str_c])
    plt.fill_between(sbp_avgs['days'], sbp_avgs[str_c] - sbp_stds[str_c], sbp_avgs[str_c] + sbp_stds[str_c], alpha=0.5)
plt.show()
```
